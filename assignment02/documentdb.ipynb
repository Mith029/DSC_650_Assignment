```python
# Importing libraries
from pathlib import Path
import json
import os
import math
from tinydb import TinyDB

# Creating files to store patient information
current_dir = Path(os.getcwd()).absolute()
results_dir = current_dir.joinpath('results')
kv_data_dir = results_dir.joinpath('kvdb')
kv_data_dir.mkdir(parents=True, exist_ok=True)


class DocumentDB(object):
    person_info = []
    def __init__(self, db_path):
        ## You can use the code from the previous exmaple if you would like
        people_json = kv_data_dir.joinpath('people.json')
        visited_json = kv_data_dir.joinpath('visited.json')
        sites_json = kv_data_dir.joinpath('sites.json')
        measurements_json = kv_data_dir.joinpath('measurements.json')

        self._db_path = Path(db_path)
        self._db = None
        
        
        # opening each json file
        with open(measurements_json,'r') as f:
            measurement_json = json.load(f)
        
        with open(people_json,'r') as f:
            people_json = json.load(f)
        
        with open(sites_json,'r') as f:
            sites_json = json.load(f)
            
        with open(visited_json,'r') as f:
            visited_json = json.load(f)
            
        # Checking for NaN value and replacing with '-'
        for key,value in visited_json.items():
            if isinstance(value['visit_date'], float) and math.isnan(value['visit_date']):
                visited_json[key]['visit_date'] = '-'
        
        # Obtaining patient information from json files
        visit_ids = []
        for per_id,per_info in people_json.items():
            temp_dic = per_info

            visits_list = []
            for measure_id,measure_info in measurement_json.items():
                visit_id = measure_info['visit_id']
                if measure_info['person_id'] == per_id:
                    for vis_id in visited_json.items():
                        site_id = vis_id[1]['site_id']
                        if visit_id == vis_id[1]['visit_id'] and visit_id not in visit_ids:
                            temp_vis_dic = vis_id[1]
                            temp_vis_dic["site"] = sites_json[site_id]
                            visits_list.append(temp_vis_dic)
                            visit_ids.append(visit_id)
                            break

                    visits_list.append(measure_info)

            dummy_list = []
            dummy_dic = {}
            dummy_measure = []
            visited_patients = []
            
            # Organizing patient information into json format to dump into database
            i = 0
            while i < len(visits_list):            
                if dummy_dic == {}:
                    if 'site' in visits_list[i]:
                        if dummy_measure == []:
                            visited_id = visits_list[i]['visit_id']
                            dummy_dic.update(visits_list[i])
                            i+= 1
                        else:
                            dummy_dic['measurements'] = dummy_measure
                            visited_patients.append(dummy_dic)
                            dummy_dic = {}
                            dummy_measure = []

                    elif dummy_measure == []:
                        visited_id = visits_list[i]['visit_id']
                        dummy_measure.append(visits_list[i])
                        i+= 1
                    elif visits_list[i]['visit_id'] == visited_id:
                        dummy_measure.append(visits_list[i])
                        i+= 1
                    else:
                        dummy_dic['measurements'] = dummy_measure
                        visited_patients.append(dummy_dic)
                        visited_id = visits_list[i]['visit_id']
                        dummy_dic = {}
                        dummy_measure = []

                else:
                    for n in range(i,len(visits_list)):
                        if visits_list[n]['visit_id'] == visited_id:
                            dummy_measure.append(visits_list[n])
                            i+=1
                            if i == len(visits_list):
                                dummy_dic['measurements'] = dummy_measure
                                visited_patients.append(dummy_dic)
                                break 
                        else:
                            dummy_dic['measurements'] = dummy_measure
                            visited_patients.append(dummy_dic)
                            dummy_dic = {}
                            dummy_measure = []
                            break

            per_info["visits"] = visited_patients
            
            self.person_info.append(per_info)
        self._load_db()

    # Loading patient information to the database
    def _load_db(self):
        self._db = TinyDB(self._db_path)
        self._db.insert_multiple(self.person_info)
        self._db.close()
```

```python
# Creating a database file
db_path = results_dir.joinpath('patient-info.json')
if db_path.exists():
    os.remove(db_path)

db = DocumentDB(db_path)
```
